Import Libraries

```python
import pandas as pd
import numpy as np
```

Load Dataset

```python
df=pd.read_csv('Clean_Dataset.csv')
print("Dataset Loaded Successfully")

df["route"] = (
    df["source_city"]
    + " → " +
    df["destination_city"]
)

df.head()
```

Executive KPI Dashboard

```python
executive_kpi = pd.DataFrame({

    "Metric":[
        "Total Flights",
        "Total Revenue",
        "Average Ticket Price",
        "Total Airlines",
        "Total Routes",
        "Average Duration"
    ],

    "Value":[
        len(df),
        df["price"].sum(),
        round(df["price"].mean(),2),
        df["airline"].nunique(),
        df["route"].nunique(),
        round(df["duration"].mean(),2)
    ]

})

executive_kpi
```

Top Revenue Airline

```python
top_airline = (
    df.groupby("airline")["price"]
    .sum()
    .idxmax()
)

print(
    "Top Revenue Airline:",
    top_airline
)
```

Most Profitable Route

```python
top_route = (
    df.groupby("route")["price"]
    .sum()
    .idxmax()
)

print(
    "Most Profitable Route:",
    top_route
)
```

Business Class Revenue %

```python
business_revenue = (
    df[df["class"]=="Business"]
    ["price"]
    .sum()
)

business_percentage = round(

    business_revenue
    /
    df["price"].sum()
    *100,

    2
)

print(
    "Business Class Revenue:",
    business_percentage,
    "%"
)
```

Final Business Recommendations

```python
recommendations = [

"Focus on high revenue routes.",

"Increase premium business class offerings.",

"Optimize pricing for last-minute bookings.",

"Expand operations on profitable routes.",

"Use dynamic pricing strategies."
]

for i,rec in enumerate(
    recommendations,
    1
):

    print(
        f"{i}. {rec}"
    )
```
